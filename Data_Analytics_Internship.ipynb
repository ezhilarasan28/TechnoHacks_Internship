### Perform Outliers for the dataset test.csv

```python
ds.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Box plot for outlier detection of 'Age' and 'Sex' columns
sns.boxplot(data=ds[['Age', 'Sex']])
plt.show()
```

```python
sns.violinplot(x='Sex', y='Age', data=ds)
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Age Distribution by Sex')
plt.show()
```

```python
sns.histplot(data=ds, x='Age', hue='Sex', kde=True)
plt.xlabel('Age')
plt.title('Age Distribution by Sex')
plt.show()
```

```python
sns.countplot(data=ds, x='Sex')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Age')
plt.show()
```
